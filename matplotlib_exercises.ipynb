```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib widget
# Apparemment, VSCode ne gère pas le mode "notebook", mais ce n'est pas interactif.
```

```python
x = np.linspace(0, 10, 1000)

fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

```python
hist1 = np.random.normal(0, 0.5, 1000)
hist2 = np.random.normal(-1.5, 1.2, 1000)
hist3 = np.random.normal(1, 2, 1000)

fig, ax = plt.subplots()
ax.hist(hist1, bins=50, alpha=0.5, density=True, histtype='stepfilled', label="ds1")
ax.hist(hist2, bins=50, alpha=0.5, density=True, histtype='stepfilled', label="ds2")
ax.hist(hist3, bins=50, alpha=0.5, density=True, histtype='stepfilled', label="ds3");

ax.legend(frameon=False);
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0,15,100)
x = np.sin(z)
y = np.cos(z)

ax.scatter3D(x, y, z, c=z, cmap='seismic')
```

```python
plt.style.use("seaborn-whitegrid")

fig, ax = plt.subplots(2, 3, sharex=True)
fig.suptitle("Typical functions")
fig.subplots_adjust( wspace=0.4)
x = np.linspace(0, 1, 100)

ax[0,0].set_xlim(0, 1)
ax[0,1].set_xlim(0, 1)
ax[0,2].set_xlim(0, 1)

ax[0,0].plot(x, x)
ax[0,1].plot(x, x**2)
ax[0,2].plot(x, x**0.5)
ax[1,0].plot(x, x)
ax[1,1].plot(x, np.log(x))
ax[1,2].plot(x, np.exp(x))

ax[0,0].set_title("$x$")
ax[0,1].set_title("$x^2$")
ax[0,2].set_title("$\sqrt{x}$")
ax[1,0].set_title("$x$")
ax[1,1].set_title("$log(x)$")
ax[1,2].set_title("$e^x$");

plt.savefig("functions.jpg")
```
